1- Usando a rede do desafio "GraphMissingEdges.gml", faça o que é pedido abaixo

a- Compute a centralidade de grau e identifique os top 50 lugares de acordo com essa métrica. Mostre o nome e categorias dos lugares.

```python
import networkx as nx

G = nx.read_gml('GraphMissingEdges.gml')

# centralidade de grau
degree_centrality = nx.degree_centrality(G)

# Classificação dos nós com base na centralidade de grau
top_places = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:50]

for node_id, centrality in top_places:
    place_name = G.nodes[node_id]['name']
    categories = G.nodes[node_id]['categories']
    print(f"Lugar: {place_name}, Categorias: {categories}, Centralidade de Grau: {centrality}")
```

a-O que acontece com as valores de PageRank dos nós conforme você aumenta o parâmetro de amortecimento (α) ?

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.read_gml('redeTRES.gml')

# parâmetro de amortecimento
alphas = [0.1, 0.3, 0.5, 0.7, 0.85, 0.9]

# valores de PageRank para cada valor de alpha
pagerank_values = {}
for alpha in alphas:
    pagerank = nx.pagerank(G, alpha=alpha)
    pagerank_values[alpha] = pagerank

for alpha, pagerank in pagerank_values.items():
    plt.plot(list(pagerank.values()), label=f'Alpha = {alpha}')

plt.xlabel('Nós')
plt.ylabel('PageRank')
plt.title('Valores de PageRank para diferentes valores de Alpha')
plt.legend()
plt.show()
```

b- O que acontece com as valores de PageRank dos nós quando o parâmetro de amortecimento (α) é 0? Explique a sua resposta

```python
import networkx as nx
import matplotlib.pyplot as plt

# Carregue o grafo a partir do arquivo GML
G = nx.read_gml('redeTRES.gml')

# Calcule o PageRank com alpha = 0
pagerank = nx.pagerank(G, alpha=0)

# Crie um gráfico de barras para exibir os valores de PageRank
plt.figure(figsize=(10, 5))
nodes = G.nodes()
pr_values = [pagerank[node] for node in nodes]
plt.bar(nodes, pr_values)
plt.xlabel('Nós')
plt.ylabel('PageRank (Alpha = 0)')
plt.title('Valores de PageRank com Alpha = 0')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

R: Quando o parâmetro de amortecimento (α) é definido como 0 no algoritmo PageRank, isso significa que não há amortecimento, e o algoritmo se comporta de maneira diferente em relação ao PageRank tradicional. Nesse caso, o PageRank não considera a probabilidade de um usuário aleatório saltar para qualquer página, sempre presumindo que o usuário continuará navegando sem parar.

3- Explique resumidamente a ideia de Hubs e Authorities no contexto de métricas de centralidade de rede.

A ideia de Hubs e Authorities refere-se a dois tipos distintos de nós em uma rede, cada um com sua própria importância relativa.

Hubs (Concentradores) são nós que têm muitos links de saída para outros nós. Eles atuam como "centros de distribuição" de informações e geralmente são nós que apontam para muitos outros na rede. Hubs tendem a ser importantes para disseminar informações para uma ampla audiência.

Authorities (Autoridades) são nós que recebem muitos links de entrada de outros nós. Eles são frequentemente vistos como especialistas ou fontes confiáveis de informações em uma determinada área. Authorities são importantes porque são referenciados e citados por outros nós da rede.

Essa ideia é frequentemente usada em algoritmos como o PageRank, onde os nós na rede são classificados com base em sua combinação de hubness (quantos links eles têm para outros) e authoritativeness (quantos links eles recebem de outros). Em resumo, Hubs disseminam informações, enquanto Authorities são reconhecidos como fontes confiáveis. Ambos desempenham papéis cruciais na disseminação e na confiabilidade da informação em uma rede.
